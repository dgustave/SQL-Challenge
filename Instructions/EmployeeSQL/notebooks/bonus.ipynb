```python
#1. Import database to pandas.
# Create string jdbc:postgresql://{host}[:{port}]/[{database}] related to DBeaver
#'postgresql+psycopg2://postgres:postgres@localhost/db_name'
def jdbc_base(database, server, host, port, user, password):
    postgres_db = {'database': f'{database}',
                   'username': f'{user}',
                   'password': f'{password}',
                   'drivername': f'{server}+psycopg2',
                   'host': f'{host}',
                   'port': port}
    return URL(**postgres_db)

url = jdbc_base('postgres', 'postgresql', "localhost", 5432, 'postgres', 'postgres')
```

```python
# Searches left-to-right, set default schemas to use, make sure it is public/default 
dbschema = 'SQLChallenge,public'
engine = create_engine(url, connect_args={'options': '-csearch_path={}'.format(dbschema)})
```

```python
# Creates String for queries: 
def load_data(schema, table):

    sql_command = "SELECT * FROM {}.{};".format(str(schema), str(table))
    print (sql_command)
    # Load the data
    data = pd.read_sql(sql_command, engine.connect())
    print(data.shape)
    return (data)
load_data('sqlchallenge', 'potential_earnings')
```

```python
#2. Create a histogram to visualize the most common salary ranges for employees.
data = load_data('sqlchallenge', 'potential_earnings')
# Find min/max for salaries and create bins for salaries
salaries = data['salary']
salaries.max(), salaries.min()
```

```python
# Plot histogram:
plt.hist(data["salary"], alpha=0.5,bins=8)
# np.histogram_bin_edges(salary, bins= salary_bins, range=None, weights=None, label=group_names)
plt.xlabel('Salary ($)')
plt.ylabel('Frequency of Salary')
plt.title('Common Salary Ranges')
plt.style.use('ggplot')
plt.savefig('../Images/Salary_Ranges_Histogram.png')
plt.show()
```

```python
# Create bins and count number of people within range of the salary: 
salary_bins = [0, 40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000]
group_names = ["<39,999", "40,000-49,999", "50,000-59,999", "60,000-69,999", "70,000-79,999", "80,000-89,000", "90,000-99,000", "100,000-119,999, 120,000+"]

# Categorize the employees using the salary bins
data["Salary Ranges"] = pd.cut(data["salary"], salary_bins, labels=group_names)

# Calculate the Numbers and Percentages by Age Group
salary_data = data["Salary Ranges"].value_counts()
s_percents = salary_data/(sum(salary_data))
sa_percents = s_percents.map("{:.2%}".format)
salary_df = pd.DataFrame({"Employees in Range": salary_data, "Salary Brackets": sa_percents })
salary_df
```

```python

#List the values of each section of the pie chart
sizes = s_percents

#Determoine which section of the circle to detach
explode =  s_percents

#Create the pie chart based upon the values 
plt.pie(sizes, explode=explode, labels=group_names, autopct="%1.1f%%", shadow=True, startangle=100)
plt.title("Salary Bracket")
#Set equal axis
plt.axis("equal")
```
